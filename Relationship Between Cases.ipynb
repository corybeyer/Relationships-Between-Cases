```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
```

```python
abc = [
    ['A', 'B', 'INC_1', 'D'],
    ['E', 'B', np.NaN, np.NaN],
    ['F', np.NaN, 'INC_1', np.NaN],
    ['G', np.NaN, np.NaN, 'D'],
    ['H', np.NaN, np.NaN, np.NaN],
    ['I', np.NaN, np.NaN, 'J'],
    ['K', np.NaN, 'INC_1', 'D'],
    ['L', 'B', np.NaN, 'J']
]
```

```python
Data = pd.DataFrame(abc, columns = ['Child', 'Parent', 'Inc', 'Grand Father'])
```

```python
G = nx.DiGraph()

for _, row in Data.iterrows():
    child = row['Child']
    for col in ['Parent', 'Inc', 'Grand Father']:
        parent = row[col]
        if pd.notna(parent):
            G.add_edge(parent, child)
```

```python
pos = nx.spring_layout(G, seed=42)  # positions for all nodes
nx.draw(G, pos, with_labels=True, node_color="lightblue", font_weight="bold", node_size=700, font_size=18)
plt.show()
```
